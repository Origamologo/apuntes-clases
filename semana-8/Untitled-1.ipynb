```python
import requests
import re
import pandas as pd
from bs4 import BeautifulSoup
```

```python
resultados_zapatillas = {"producto":[], 
                        "precio_rebajado": [], 
                        "precio_original": [], 
                         "descuento": [], 
                        "enlaces": []}
```

```python
url = "https://deportesmoya.es/1585-zapatillas-baloncesto"
```

```python
resq1 = requests.get(url)
```

```python
resq1.status_code
```

```python
resq1
```

```python
# Esto te descarga to el html, es un infierno
# resq1.content
```

```python
soup = BeautifulSoup(resq1.content, "html.parser")
```

```python
soup
```

```python
productos = soup.find_all("h3",{"class": "h3 product-title"})
```

```python
len(productos)
```

```python
productos[0]
```

```python
productos[0].getText()
```

```python
dir(productos)
```

```python
for a in productos:
    resultados_zapatillas['producto'].append(a.getText())
```

```python
precios = soup.find_all('span', {'class': 'price'})
```

```python
len(precios)
```

```python
precios[0]
```

```python
precios[9].getText()
```

```python
patron = '\d+,\d+'
```

```python
for precio in precios:
    resultados_zapatillas['precio_rebajado'].append(re.findall(patron, precio.getText())[0])
```

```python
resultados_zapatillas
```

```python
# Esto es exactamente lo mismok que el find_all
prueba = soup.select('span.price')
```

```python
prueba
```

```python
precios_originales = soup.select('span.regular-price')
```

```python
for precio in precios_originales:
    resultados_zapatillas['precio_original'].append(re.findall(patron, precio.getText())[0])
```

```python
resultados_zapatillas
```

```python
enlaces = soup.find_all('div', {'class': 'thumbnail-container'})
```

```python
enlaces[0].getText()
```

```python
enlaces[0].find('a').get('href')
```

```python

for enlace in enlaces:
    resultados_zapatillas['enlaces'].append(enlace.find('a').get('href'))
```

```python
descuentos = soup.find_all('li', {'class': 'product-flag discount'})
```

```python
descuentos[:2]
```

```python
for descuento in descuentos:
    resultados_zapatillas['descuento'].append(descuento.getText())
```

```python
df_zapatillas = pd.DataFrame(resultados_zapatillas)
```
